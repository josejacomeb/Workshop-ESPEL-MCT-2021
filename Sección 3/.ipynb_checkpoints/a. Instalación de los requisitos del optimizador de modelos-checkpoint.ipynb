# Instalación de los requisitos del optimizador de modelos de Aprendizaje Profundo

## Introducción

Es una línea de comandos que funciona en varias plataformas que facilita la transición entre el modelo de entrenamiento y el ambiente final del lanzamiento del proyecto, el optimizador de modelos es compatible con varios de los formatos populares de Deep Learning como: Caffe, TensorFlow, Kaldi, MXNet o modelos de AI estándar en el formato ONNX[1].

El objetivo de esto, llevar acabo esto la transformaci'on de esos modelos hacia un formato estándar de Intel llamado la Representación Intermedia(IR por sus siglas en Ingles), dicho modelo que ha sido producido por el Optimizador de Modelos es Compatible con el Motor de Inferencia de Intel[2].

Básicamente, el optimizador de modelos, convierte modelos pre-entrenados hacia dos archivos a saber:

![Deep Learning Model Optimizer](https://docs.openvinotoolkit.org/latest/workflow_steps.png)
<div align="center"><b>Fig 1.</b> Aplicación del Optimizador de Modelos para procesar un modelo pre-entrenado [1]</div>

La representación intermedia, es un par de archivos que describe la red
- `.xml` Describe la topología de la red
- `.bin` Contiene los pesos y los biases en datos binarios

## Instalación de los pre-requisitos del Optimizador de Modelos
Cabe mencionar que para poder usar el optimizador de modelos, es necesario primero instalar las dependencias de cada Framework, afortunadamente la instalación de estos es de manera sencilla ya que vienen convenientemente con scripts de Instalación para proveer compatibilidad con los principales frameworks de Inteligencia Artificial soportados por la actual distribución de OpenVINO, el usuario es libre de instalar el paquete que le convenga, pero para este curso es necesario al menos instalar **la compatibilidad para un Framework** que se presentan a continuación[3].

Primero que todo, hay que abrir un terminal o CMD y cambiarse nuestra ubicación hacia el Directorio de Scripts Instalación de Soportde para el Optimizador de Modelos de OpenVINO, normalmente es el siguiente:

- *Windows*
```bash
cd C:\Program Files (x86)\Intel\openvino_2021\deployment_tools\model_optimizer\install_prerequisites\
```
- *Linux*
```bash
cd /opt/intel/openvino_2021/deployment_tools/model_optimizer/install_prerequisites/
```
Luego es necesario activar el entorno del OpenVINO para evitar que descargue el Kit de Herramientas por pip.

- *Windows*
```bash
"C:\Program Files (x86)\Intel\openvino_2021\bin\setupvars.bat"
```

- *Linux*

```bash
source /opt/intel/openvino_2021/bin/setupvars.sh
```

### Instalación del Paquete Completo
El paquete completo contiene compatibilidad para todas los Frameworks soportados por OpenVINO y usualmente es el más pesado de todos. Una vez ingresada a la carpeta de Scripts de instalación como en el paso anterior se procede con el siguiente comando:
- *Windows*
```bash
install_prerequisites.bat
```
- *Linux*
```bash
sudo ./install_prerequisites.sh
```

![OpenVINO Demo](Imágenes/openvino_mo_todos.png)
<div align="center"><b>Fig 2.</b> Instalación del Paquete Completo del Optimizador de Modelos en Linux </div>

El programa empezará a descargar las dependencias  correspondientes y al final aparecerá un mensaje del siguiente estilo, que mostrará que la instalación ha sido satisfactoria. 

```bash
        - Inference Engine found in:    /opt/intel/openvino_2021.4.582/python/python3.8/openvino
Inference Engine version:       2021.4.0-3839-cd81789d294-releases/2021/4
Model Optimizer version:        2021.4.0-3839-cd81789d294-releases/2021/4
[WARNING] All Model Optimizer dependencies are installed globally.
[WARNING] If you want to keep Model Optimizer in separate sandbox
[WARNING] run install_prerequisites.sh "{caffe|tf|tf2|mxnet|kaldi|onnx}" venv

```

### Instalación del Paquete para Tensorflow 2
Este paquete contiene la compatibilidad para las Librerías TensorFlow v2 y v1, en este presente documento se mostrará la instalación para el Paquete 2, para la instalación en la versión 1, por favor referirse a la referencia[3], los comandos son los siguientes: 
- *Windows*
```bash
install_prerequisites_tf2.bat
```
- *Linux*
```bash
sudo ./install_prerequisites_tf2.sh
```

### Instalación del Paquete para Caffe
La instalación del Paquete de Compatibilidad de Caffe corresponde al siguiente comando: 
- *Windows*
```bash
install_prerequisites_caffe.bat
```
- *Linux*
```bash
sudo ./install_prerequisites_caffe.sh
```

### Instalación del Paquete para ONNX
El paquete de ONNX se puede instalar con el siguiente comando:

- *Windows*
```bash
install_prerequisites_onnx.bat
```
- *Linux*
```bash
sudo ./install_prerequisites_onnx.sh
```
