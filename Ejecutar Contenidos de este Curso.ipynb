# ¿Cómo ejecutar el Contenido de este Curso?
Como se pudo notar, el contenido mostrado en la aula virtual de Moodle es estático, sin embargo el contenido de este Workshop ha sido diseñado para que se sea interactivo, por lo que para ejecutar el contenido de este curso, a parte de la Instalación del Software OpenVINO, es necesario realizar estos dos procedimientos

## Pre-requisitos
1. Descargar el código de este Workshop en su computador personal
2. Ejecutar la interfaz de Jupyter Notebook. 

## Descargar el código de este Workshop en su computador personal
El contenido de este Workshop, está alojado de manera pública en el Servidor de Github, el cual es accesible con el siguiente link: [https://github.com/josejacomeb/Workshop-ESPEL-MCT-2021](https://github.com/josejacomeb/Workshop-ESPEL-MCT-2021). De ahí tenemos dos opciones a saber para descargar el código:

### Clonar el código a través de la aplicación de Git
La manera más eficiente de realizar este proceso es a través de la aplicación gráfica de [Git](https://git-scm.com/download/win) o [Github](https://desktop.github.com/). Una vez clickeado en el Botón de **Código** se despliegan las opciones para  clonar, se sugiere la opción *HTTPS*. Es necesario instalar esas aplicaciones, si se ha optado por la aplicación *Git*, con el siguiente código de ejemplo se puede clonar el repositorio como HTTPS.
```bash
cd /home/josejacomeb/Documents/ #Nombre de la Carpeta donde se desea clonar los archivos
git clone https://github.com/josejacomeb/Workshop-ESPEL-MCT-2021.git
```
El código sigujiente Más información de como [Clonar un repositorio](https://docs.github.com/es/github/creating-cloning-and-archiving-repositories/cloning-a-repository-from-github/cloning-a-repository) en dicho link.

### Descargar el código en Formato ZIP
Esta opción no necesita de ningún programa, a parte de un programa para descomprimir los archivos. Una vez dado click en la opción **Code**, es necesario dar click en la Opción *Download ZIP*, descomprimir el código en la carpeta deseada.

## Ejecutar la interfaz de Jupyter Notebook. 

Una vez descargado y descomprimido el código, es necesario leerlo y ejecutarlo, para ello primero es necesario instalar la Interfaz de Jupyter Notebook.

### Instalación

Una vez instalado Python en el computador, solo es necesario instalar JupyterLab a través de la aplicación PIP(que viene incluída con Python), copiar y ejecutar el código en CMD en Windows o en el Terminal en Linux.
- **Windows**
```cmd
pip install jupyterlab
```
- **Linux**
```bash
sudo pip3 install jupyterlab
```

### Iniciar la interfaz de Jupyter Notebook

Una vez finalizada la instalación del anterior paso, basta en ejecutar el Jupyter Notebook a través del CMD de Windows o el Terminal de Linux, cabe destacar que el Jupyter Notebook debe ser ejecutado dentro de la Carpeta de Códigos del Presente Workshop. Un ejemplo es el siguiente.
```bash
cd /home/josejacomeb/Documents/Workshop-ESPEL-MCT-2021/
jupyter notebook
```
Solo hace falta ejecutar el comando `jupyter notebook` y entonces se despliega algunos mensajes informativos, como el URL local para acceder al Notebook, como se puede ver en la siguiente imágen.

### Ejecución de Celdas
Una vez ingresada a los archivos del curso, se puede notar que cualquier documento está dividido en celdas, cada celda puede ser ejecutada, las celdas que el instructor marque como **Ejecutable** deberán seguir ese procedimiento.  
El ejecutar una celda es fácil, basta con *Seleccionar la Celda* y luego dar click en el Botón **Run** de la parte superior central o a su vez *Seleccionar la Celda* y utilizar el atajo de teclado `Mayus + Enter`.
